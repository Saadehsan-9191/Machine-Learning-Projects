### Convolutional Neural Network (CNN) - MNIST Handwritten Digit Classification

1. Import libraries and download dataset 

```python
# Import the necessary libraries
from keras.datasets import mnist       # to load the MNIST dataset
from keras.models import Sequential    # to build a sequential CNN model
from keras.layers import Conv2D        # convolutional layer
from keras.layers import MaxPooling2D  # max pooling layer
from keras.layers import Flatten       # flattening layer
from keras.layers import Dense         # fully connected layers


# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()  # loads training and testing data

# Print the shapes of the datasets to understand their structure
print("Training images shape:", x_train.shape)   # should be (60000, 28, 28)
print("Training labels shape:", y_train.shape)   # should be (60000,)
print("Testing images shape:", x_test.shape)     # should be (10000, 28, 28)
print("Testing labels shape:", y_test.shape)     # should be (10000,)


# Visualize a sample image from the dataset
plt.imshow(x_train[0], cmap='gray')    # show the first training image in grayscale
plt.title(f"Label: {y_train[0]}")      # display the corresponding label
plt.axis('off')                        # hide axis for cleaner view
plt.show()                             # render the image
```

2. Reshaping and Normalizing the Dataset

```python
# Reshape and Normalize the Dataset
# Reshape the training images to add the channel dimension (1 for grayscale)
# New shape becomes: (60000, 28, 28, 1)
x_train = x_train.reshape(60000, 28, 28, 1) / 255.0   # divide by 255 to normalize pixel values

# Reshape the testing images to add the channel dimension
# New shape becomes: (10000, 28, 28, 1)
x_test = x_test.reshape(10000, 28, 28, 1) / 255.0     # normalize test images as well

# Print the new shapes to confirm the transformation
print("New training images shape:", x_train.shape)    # should be (60000, 28, 28, 1)
print("New testing images shape:", x_test.shape)      # should be (10000, 28, 28, 1)
```

3. Model

```python
# Build the Convolutional Neural Network (CNN)

# Create a Sequential model (layers will be added one after another)
model = Sequential()   # start an empty model

# Add the first layer: Convolutional Layer
model.add(
    Conv2D(
        filters=28,              # number of filters (feature detectors)
        kernel_size=(3, 3),      # size of each filter (3x3 window)
        activation='relu',       # activation function to introduce non-linearity
        input_shape=(28, 28, 1)  # shape of each input image (height, width, channels)
    )
)

# Add the second layer: MaxPooling Layer
model.add(
    MaxPooling2D(
        pool_size=(2, 2)         # reduces each feature map by taking max in 2x2 blocks
    )
)

# Add the third layer: Flatten Layer
model.add(
    Flatten()                    # converts 2D feature maps into a 1D vector
)

# Add the fourth layer: Dense (Fully Connected) Layer
model.add(
    Dense(
        units=128,               # number of neurons in the dense layer
        activation='relu'        # activation function
    )
)

# Add the output layer: Dense Layer with Softmax
model.add(
    Dense(
        units=10,                # 10 classes (digits 0вЂ“9)
        activation='softmax'     # softmax outputs probabilities that sum to 1
    )
)

# Compile the model with chosen hyperparameters
model.compile(
    optimizer='adam',                     # optimizer for training
    loss='sparse_categorical_crossentropy',  # loss function for multi-class classification
    metrics=['accuracy']                  # track accuracy during training
)

# Print the model summary to see the architecture
model.summary()   # shows each layer, output shape, and number of parameters
```

4. Training the Model and Evaluation

```python
# Train the CNN model
# Train the model using the training data
history = model.fit(
    x_train,            # training images
    y_train,            # training labels
    epochs=10,         # number of times the model sees the full dataset
    validation_data=(x_test, y_test)  # evaluate on test data during training
)


# Evaluate the model on the test dataset
test_loss, test_accuracy = model.evaluate(
    x_test,     # test images
    y_test      # test labels
)

# Print the final test results
print("Final Test Loss:", test_loss)          # how wrong the model is
print("Final Test Accuracy:", test_accuracy)  # how many test images it got right
```
